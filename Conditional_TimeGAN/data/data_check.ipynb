```python
import csv
import numpy as np
import matplotlib.pyplot as plt

category = ['JK', 'JP', 'BL','VT','SL','SK','OP','TC','CT','KT','KC','TS','PD']

for cat in category:
    # 빈 리스트 생성
    sales_quantity = []

    # CSV 파일 경로
    csv_file_path = f'/home/langleman/TimeGAN4paper/data/{cat}_weekly_new_train.csv'  # 실제 파일 경로로 변경하세요.

    # CSV 파일 읽기
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        # 각 행에 대해 반복하여 리스트에 추가
        for row in reader:
            # 첫 번째 열에 있는 값을 부동소수점으로 변환하여 리스트에 추가
            sales_quantity.append(float(row[0]))

    # Min-Max Scaling을 적용하여 sales_quantity를 [0, 1] 범위로 정규화
    min_val = min(sales_quantity)
    max_val = max(sales_quantity)
    scaled_sales_quantity = [(x - min_val) / (max_val - min_val) for x in sales_quantity]

    # 데이터를 나눌 구간의 길이
    interval_length = 4

    # 구간별 분산을 저장할 리스트
    variances = []

    # 데이터를 구간으로 나누어 각 구간의 분산을 계산
    for i in range(0, len(scaled_sales_quantity)):
        interval_data = scaled_sales_quantity[i:i+interval_length]
        interval_variance = np.var(interval_data)
        variances.append(interval_variance)

    # 시간에 따른 분산의 변화 시각화
    plt.plot(variances)
    plt.xlabel('Time')
    plt.ylabel('Variance')
    plt.title(f'Variance over Time - {cat}')
    plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

category = ['JK', 'JP', 'BL','VT','SL','SK','OP','TC','CT','KT','KC','TS','PD']

# 전체 카테고리를 3개씩 나열하여 플롯을 생성
num_rows = len(category) // 3 + (len(category) % 3 > 0)  # 행의 개수
fig, axs = plt.subplots(num_rows, 3, figsize=(15, 5*num_rows))

# 각 카테고리별로 푸리에 변환을 수행하고 서브 플롯에 표시
for i, cat in enumerate(category):
    # 빈 리스트 생성
    sales_quantity = []

    # CSV 파일 경로
    csv_file_path = f'/home/langleman/TimeGAN4paper/data/{cat}_weekly_new_train.csv'  # 실제 파일 경로로 변경하세요.

    # CSV 파일 읽기
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        # 각 행에 대해 반복하여 리스트에 추가
        for row in reader:
            # 첫 번째 열에 있는 값을 부동소수점으로 변환하여 리스트에 추가
            sales_quantity.append(float(row[0]))

    # 시간 간격 설정
    time_interval = 1  # 시간 간격은 1로 설정됨(가정)

    # 푸리에 변환 계산
    fft_result = np.fft.fft(sales_quantity)

    # 주파수 배열 생성
    N = len(sales_quantity)
    frequencies = np.fft.fftfreq(N, d=time_interval)

    # 결과 시각화
    row_index = i // 3
    col_index = i % 3
    axs[row_index, col_index].plot(frequencies[:N//2], np.abs(fft_result[:N//2]))  # 양측 대칭성을 고려하여 절반만 표시
    axs[row_index, col_index].set_xlabel('Frequency (Hz)')
    axs[row_index, col_index].set_ylabel('Amplitude')
    axs[row_index, col_index].set_title(f'Fourier Transform of sales_quantity - {cat}')
    axs[row_index, col_index].grid()

# 마지막 미사용 서브 플롯 삭제
if len(category) % 3 != 0:
    for j in range(len(category) % 3, 3):
        fig.delaxes(axs[num_rows - 1, j])

plt.tight_layout()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

category = ['JK', 'JP', 'BL','VT','SL','SK','OP','TC','CT','KT','KC','TS','PD']

# 각 카테고리별로 ACF 및 PACF를 계산하고 표시
for cat in category:
    # 빈 리스트 생성
    sales_quantity = []

    # CSV 파일 경로
    csv_file_path = f'/home/langleman/TimeGAN4paper/data/{cat}_weekly_new_train.csv'  # 실제 파일 경로로 변경하세요.

    # CSV 파일 읽기
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        # 각 행에 대해 반복하여 리스트에 추가
        for row in reader:
            # 첫 번째 열에 있는 값을 부동소수점으로 변환하여 리스트에 추가
            sales_quantity.append(float(row[0]))

    # ACF 및 PACF 계산 및 시각화
    plt.figure(figsize=(6, 3))
    ax1 = plt.subplot(121)
    plot_acf(sales_quantity, lags=10, ax=ax1)
    ax1.set_title(f'ACF - {cat}')

    ax2 = plt.subplot(122)
    plot_pacf(sales_quantity, lags=10, ax=ax2)
    ax2.set_title(f'PACF - {cat}')

    plt.tight_layout()
    plt.show()

```
