```python
import csv
import numpy as np
import matplotlib.pyplot as plt

category = ['JK', 'JP', 'BL','VT','SL','SK','OP','TC','CT','KT','KC','TS','PD']

for cat in category:
    #   
    sales_quantity = []

    # CSV  
    csv_file_path = f'/home/langleman/TimeGAN4paper/data/{cat}_weekly_new_train.csv'  #    .

    # CSV  
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        #      
        for row in reader:
            #         
            sales_quantity.append(float(row[0]))

    # Min-Max Scaling  sales_quantity [0, 1]  
    min_val = min(sales_quantity)
    max_val = max(sales_quantity)
    scaled_sales_quantity = [(x - min_val) / (max_val - min_val) for x in sales_quantity]

    #    
    interval_length = 4

    #    
    variances = []

    #       
    for i in range(0, len(scaled_sales_quantity)):
        interval_data = scaled_sales_quantity[i:i+interval_length]
        interval_variance = np.var(interval_data)
        variances.append(interval_variance)

    #     
    plt.plot(variances)
    plt.xlabel('Time')
    plt.ylabel('Variance')
    plt.title(f'Variance over Time - {cat}')
    plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

category = ['JK', 'JP', 'BL','VT','SL','SK','OP','TC','CT','KT','KC','TS','PD']

#   3   
num_rows = len(category) // 3 + (len(category) % 3 > 0)  #  
fig, axs = plt.subplots(num_rows, 3, figsize=(15, 5*num_rows))

#        
for i, cat in enumerate(category):
    #   
    sales_quantity = []

    # CSV  
    csv_file_path = f'/home/langleman/TimeGAN4paper/data/{cat}_weekly_new_train.csv'  #    .

    # CSV  
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        #      
        for row in reader:
            #         
            sales_quantity.append(float(row[0]))

    #   
    time_interval = 1  #   1 ()

    #   
    fft_result = np.fft.fft(sales_quantity)

    #   
    N = len(sales_quantity)
    frequencies = np.fft.fftfreq(N, d=time_interval)

    #  
    row_index = i // 3
    col_index = i % 3
    axs[row_index, col_index].plot(frequencies[:N//2], np.abs(fft_result[:N//2]))  #     
    axs[row_index, col_index].set_xlabel('Frequency (Hz)')
    axs[row_index, col_index].set_ylabel('Amplitude')
    axs[row_index, col_index].set_title(f'Fourier Transform of sales_quantity - {cat}')
    axs[row_index, col_index].grid()

#     
if len(category) % 3 != 0:
    for j in range(len(category) % 3, 3):
        fig.delaxes(axs[num_rows - 1, j])

plt.tight_layout()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

category = ['JK', 'JP', 'BL','VT','SL','SK','OP','TC','CT','KT','KC','TS','PD']

#   ACF  PACF  
for cat in category:
    #   
    sales_quantity = []

    # CSV  
    csv_file_path = f'/home/langleman/TimeGAN4paper/data/{cat}_weekly_new_train.csv'  #    .

    # CSV  
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        #      
        for row in reader:
            #         
            sales_quantity.append(float(row[0]))

    # ACF  PACF   
    plt.figure(figsize=(6, 3))
    ax1 = plt.subplot(121)
    plot_acf(sales_quantity, lags=10, ax=ax1)
    ax1.set_title(f'ACF - {cat}')

    ax2 = plt.subplot(122)
    plot_pacf(sales_quantity, lags=10, ax=ax2)
    ax2.set_title(f'PACF - {cat}')

    plt.tight_layout()
    plt.show()

```
