```python
import numpy as np
from scipy import stats

# 데이터 증강 전후의 SMAPE 값
smape_before = np.array([21.05878, 27.651, 16.67584, 31.8013, 13.49323, 16.39029, 12.86339, 71.64347, 32.19932, 14.57134, 28.19787, 21.0585, 44.36297])
smape_after = np.array([16.08893, 23.06576, 13.46687, 27.14261, 15.18836, 18.03054, 11.81428, 65.76128, 39.72802, 14.30337, 14.99501, 18.73198, 44.63973])

# 정규성 검정 (Shapiro-Wilk Test)
stat_before, p_before = stats.shapiro(smape_before)
stat_after, p_after = stats.shapiro(smape_after)

if p_before > 0.05 and p_after > 0.05:
    print('y')
    # 대응 표본 t-검정 (정규성 가정)
    t_stat, p_value = stats.ttest_rel(smape_before, smape_after)
else:
    print('e')
    # 윌콕슨 부호 순위 검정 (비모수 검정)
    t_stat, p_value = stats.wilcoxon(smape_before, smape_after)

print(f'p-value: {p_value}')
```

```python
import numpy as np

# 시계열 데이터 예시
time_series_data = np.array([1, 2, 3, 4, 5])

# 가우시안 노이즈 생성 (평균 0, 표준편차 0.1)
noise = np.random.normal(0, 0.1, size=time_series_data.shape)

# 노이즈 주입된 시계열 데이터
noisy_time_series_data = time_series_data + noise
```

```python
import matplotlib.pyplot as plt

plt.plot(time_series_data, label="Original")
plt.plot(noisy_time_series_data, label="Noisy")
plt.legend()
plt.show()
```
